14. Display 10 Most Popular Courses As Per Number of Subscribers

```python
top_10=df.sort_values(by='num_subscribers',ascending=False).head(10)
top_10
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter top 10
top_10 = df.sort_values(by='num_subscribers', ascending=False).head(10)

plt.figure(figsize=(10,6))  # bigger width so text fits

sns.barplot(x='num_subscribers', y='course_title', data=top_10, palette='Blues_r')

plt.title('Top 10 Most Popular Courses As Per Number of Subscribers')
plt.xlabel('Number of Subscribers')
plt.ylabel('Course Title')
plt.tight_layout()   # adjusts spacing
plt.show()
```

```python
# per different colours
plt.figure(figsize=(12,6))

sns.barplot(
    x='num_subscribers',
    y='course_title',
    data=top_10,
    palette=sns.color_palette('tab10', 10)  # 10 distinct colors
)

plt.title('Top 10 Most Popular Courses As Per Number of Subscribers')
plt.xlabel('Number of Subscribers')
plt.ylabel('Course Title')
plt.tight_layout()
plt.show()
```

15. Find The Course Which Is Having The Highest Number of Reviews.

```python
# Group by subject and sum the number of reviews
subject_reviews = df.groupby('subject')['num_reviews'].sum()

# Find the subject with the maximum reviews
top_subject = subject_reviews.idxmax()
top_reviews = subject_reviews.max()

print(f"Subject with highest reviews: {top_subject} ({top_reviews} reviews)")
```

```python
top_course = df[df['num_reviews'] == df['num_reviews'].max()]
print(top_course)
```

```python
# barplot visual
sns.barplot(x='subject',y='num_reviews',data=df)
```

```python
# to visual neatly
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(10,6))

# Create barplot with palette for different colors
sns.barplot(
    x='subject',
    y='num_reviews',
    data=df,
    palette='Set3'  # You can use any seaborn palette or list of colors
)

# Add title and labels
plt.title('Number of Reviews per Subject', fontsize=16)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)

# Rotate x-axis labels if they overlap
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```
